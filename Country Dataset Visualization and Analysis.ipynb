```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

##Data Ingestion
countries = pd.read_csv('~/projects/datasets/Country Data/top_six_economies.csv')
countries
```

```python
alcohol_abuse = pd.read_csv('~/projects/datasets/Country Data/alcoholSubstanceAbuse.csv')
alcohol_abuse
```

```python
life_expectancy = pd.read_csv('~/projects/datasets/Country Data/HALElifeExpectancyAtBirth.csv')
life_expectancy
```

```python
suicide_rate = pd.read_csv('~/projects/datasets/Country Data/crudeSuicideRates.csv')
suicide_rate
```

```python
## Data Handling
# Now let's rename all of the columns so we can merge the dataframes
suicide_rate.rename(columns={'Dim1':'Sex', 'First Tooltip': 'Crude Suicide Rates (per 100,000 population)'}, inplace=True)
life_expectancy.rename(columns={'Dim1':'Sex', 'First Tooltip': 'Healthy Life Expectancy (years)'}, inplace=True)
alcohol_abuse.rename(columns={'Dim1':'Sex', 'First Tooltip': 'Alcohol per Capita Consumption'}, inplace=True)
print(suicide_rate.columns, life_expectancy.columns, alcohol_abuse.columns, sep = '\n')
```

```python
#Now that all data sets are loaded we need to merge them
alcohol_abuse['Location'].replace(to_replace=['United States of America', 'United Kingdom of Great Britain and Northern Ireland'],
                                  value=['United States', 'United Kingdom'], inplace=True)
x = pd.merge(countries, alcohol_abuse, left_on=['Country Name', 'Year'], right_on=['Location','Period'])
life_expectancy['Location'].replace(to_replace=['United States of America', 'United Kingdom of Great Britain and Northern Ireland'],
                                  value=['United States', 'United Kingdom'], inplace=True)
dataset = pd.merge(x, life_expectancy, left_on=['Country Name', 'Year', 'Sex'], right_on=['Location','Period', 'Sex'])
suicide_rate['Location'].replace(to_replace=['United States of America', 'United Kingdom of Great Britain and Northern Ireland'],
                                  value=['United States', 'United Kingdom'], inplace=True)
top_countries = pd.merge(dataset, suicide_rate, left_on=['Country Name', 'Year', 'Sex'], right_on=['Location','Period', 'Sex'])
top_countries
```

```python
#Check for null values and duplicates
top_countries.isnull().sum()
```

```python
#Dropping Poverty and government debt columns, we will not need them
top_countries.drop(columns=['Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population)',
                           'Central government debt, total (% of GDP)'], inplace = True)
top_countries.isnull().sum()
```

```python
#Check for duplicates
top_countries.duplicated().sum()
#There are none
```

```python
#Delete all unneeded columns for top_countries
unneeded = ['Unnamed: 0', 'Location_x', 'Period_x', 'Indicator_x', 'Location_y', 'Period_y', 'Indicator_y', 
           'Location', 'Period', 'Indicator']
top_countries.drop(columns=unneeded, inplace=True)
top_countries.columns
```

```python
#Which country has the highest amount of alcohol consumption?
top_countries_genderless = top_countries[top_countries['Sex']=='Both sexes']
alcohol_grouped = top_countries_genderless[['Alcohol per Capita Consumption']].groupby(top_countries['Country Name']).mean()
alcohol_grouped.sort_values(by='Alcohol per Capita Consumption', ascending=False)
#Germany has the highest rate of alcohol consumption on average
```

```python
alcohol_grouped = top_countries_genderless['Alcohol per Capita Consumption'].groupby([top_countries_genderless['Country Name'], 
                                                                                        top_countries_genderless['Year']]).mean()
alcohol_grouped.plot(kind = 'barh', legend=None)
plt.xlabel('Alcohol Per Capita Consumption')
plt.tight_layout()
#Germany in 2000 had the highest rate of alcohol per capita for a year
```

```python
#Which sex had the highest rate of alcohol consumption?
top_countries_gendered = top_countries[top_countries['Sex']!='Both sexes']
alcohol_by_sex = top_countries_gendered['Alcohol per Capita Consumption'].groupby(top_countries_gendered['Sex']).mean()
plt.figure()
alcohol_by_sex.plot(kind='bar')
plt.ylabel('Alcohol per Capita Consumption')
plt.tight_layout()
```

```python
#What is the relationship between Alcohol per Capita Consumption over time?
plt.figure()
sns.lineplot(data=top_countries_genderless, x='Year', y='Alcohol per Capita Consumption', hue='Country Name')
#Alcohol per Capita Consumption seems to be declining in most countries except for China and India
```

```python
#Is there a correlation between life expectancy in years vs. healthy life expectancy?
plt.figure()
sns.regplot(top_countries_genderless['Healthy Life Expectancy (years)'], 
            top_countries_genderless['Life expectancy at birth, total (years)'])
print(top_countries_genderless['Healthy Life Expectancy (years)'].corr(top_countries_genderless['Life expectancy at birth, total (years)']))
#As expected there is a very strong relationship between life expectancy and healthy life expectancy 
#correlation of 0.9768
```

```python
#What is the relationship between Alcohol per Capita Consumption and Healthy life expectancy?
plt.figure()
sns.regplot(top_countries_genderless['Healthy Life Expectancy (years)'], 
            top_countries_genderless['Alcohol per Capita Consumption'])
print(top_countries_genderless['Healthy Life Expectancy (years)'].corr(top_countries_genderless['Alcohol per Capita Consumption']))
#There is still a correlation but a much weaker one of 0.634 which is not expected since the opposite trend
#would be expected (longer healthy lives mean less drinking)
```

```python
#If you are a female do you have a higher chance of having a higher healthy life expectancy?
plt.figure()
sns.lineplot(data=top_countries_gendered, x='Year', y='Healthy Life Expectancy (years)', hue='Sex')
#Females overall have a higher healthy life expectancy than men
```

```python
#Is there any year where females lived shorter than men in each of the countries?
pivoted = pd.pivot_table(top_countries_gendered, values='Healthy Life Expectancy (years)', 
                         index=['Country Name', 'Year'], columns='Sex', aggfunc='mean')
pivoted
# No, every year females had a higher healthy life expectancy 
```

```python
# Is there a correlation between Unemployment and Suicide Rates
plt.figure()
sns.regplot(top_countries_genderless['Unemployment, total (% of total labor force) (modeled ILO estimate)'],
           top_countries_genderless['Crude Suicide Rates (per 100,000 population)'])
print(top_countries_genderless['Unemployment, total (% of total labor force) (modeled ILO estimate)'].corr(
    top_countries_genderless['Crude Suicide Rates (per 100,000 population)']))
#There does not seem to be a relation between suicide rates and unemployment (very small negative correlation)
```

```python
# Are suicides higher for men or woman and in what country?
suicide_country_sex= top_countries_gendered['Crude Suicide Rates (per 100,000 population)'].groupby([top_countries_gendered['Country Name'],
                                                                               top_countries_gendered['Sex']]).mean()
reformatted = pd.DataFrame(suicide_country_sex.reset_index())
plt.figure()
sns.barplot(x="Country Name", y="Crude Suicide Rates (per 100,000 population)", hue='Sex', data=reformatted)
#Men in Japan have the highest suicide rate by far
```

```python
#What is the most common crude suicide rate across all the countries
print(top_countries_genderless['Crude Suicide Rates (per 100,000 population)'].mean(),
      top_countries_genderless['Crude Suicide Rates (per 100,000 population)'].std(),
      sep=', ')
plt.figure()
sns.distplot(top_countries_genderless['Crude Suicide Rates (per 100,000 population)'])
```

```python
#How many of those suicides are by men vs. woman
suicide_by_sex = pd.pivot_table(data=top_countries_gendered, index='Sex', columns='Country Name', 
               values='Crude Suicide Rates (per 100,000 population)', aggfunc='mean')
fig = suicide_by_sex.plot(kind="pie", subplots=True, layout= (2,3), legend=False, label="")
plt.tight_layout()
#The only country that has a higher female suicide rate than male is in China. India comes second
```

```python
#Let's look at how population growth and suicide rates are related
print(top_countries_genderless['Population growth (annual %)'].corr(top_countries_genderless['Crude Suicide Rates (per 100,000 population)']))
plt.figure()
sns.scatterplot(x='Population growth (annual %)', y ='Crude Suicide Rates (per 100,000 population)', 
                data=top_countries_gendered, hue='Sex')
#Suicide rates seem to be decreasing slightly as population growth accelerates
```

```python
#Finally what is the relationship between all the variables including GDP per capita, suicide rate, unemployment,
#healthy life expectancy, and alcoholism
pairplot = sns.pairplot(top_countries_gendered[['GDP per capita (current US$)', 'Crude Suicide Rates (per 100,000 population)',
                                       'Healthy Life Expectancy (years)', 'Unemployment, total (% of total labor force) (modeled ILO estimate)',
                                      'Alcohol per Capita Consumption', 'Sex']], hue = 'Sex')
labels = ['GDP per capita', 'Suicide rate', 'Healthy life expectancy', 'Unemployment', 'ALcohol consumption']
#There does not seem to be that meaningful of a relationship between any of the variables
```
